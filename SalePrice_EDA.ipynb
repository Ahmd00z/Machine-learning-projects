### **MODEL**

```python
from sklearn.model_selection import train_test_split , GridSearchCV
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.neighbors import KNeighborsRegressor
from xgboost import XGBRegressor
from sklearn.metrics import r2_score
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import  OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
```

```python
X = data.drop('SalePrice', axis=1)
y = data['SalePrice']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
categorical_cols = X_train.select_dtypes(include=["object"]).columns
numeric_cols = X_train.select_dtypes(include=["int64", "float64"]).columns
```

```python
preprocessor = ColumnTransformer(
    transformers=[
        ("num", StandardScaler(), numeric_cols),
        ("cat", OneHotEncoder(handle_unknown="ignore"), categorical_cols)
    ]
)
```

```python
X_train_processed = preprocessor.fit_transform(X_train)
X_test_processed = preprocessor.transform(X_test)

print("Train shape after preprocessing:", X_train_processed.shape)
print("Test shape after preprocessing:", X_test_processed.shape)
```

```python
linear_model = LinearRegression()
linear_model.fit(X_train_processed, y_train)

y_pred_linear = linear_model.predict(X_test_processed)

r2_linear = r2_score(y_test, y_pred_linear)
print("RÂ² score (Linear Regression):", r2_linear)
```

```python
# 1. Polynomial Linear Regression
poly_pipeline = Pipeline([
    ("preprocess", preprocessor),
    ("poly", PolynomialFeatures(degree=5, include_bias=False)),
    ("model", LinearRegression())
])

# 2. KNN
knn_pipeline = Pipeline([
    ("preprocess", preprocessor),
    ("model", KNeighborsRegressor())
])

# 3. XGBoost
xgb_pipeline = Pipeline([
    ("preprocess", preprocessor),
    ("model", XGBRegressor(objective="reg:squarederror", random_state=42))
])

# =====================
# Define Parameter Grids
# =====================
param_grid = {
    "poly": {
        "poly__degree": [1, 2]
    },
    "knn": {
        "model__n_neighbors": [3, 5, 7],
        "model__weights": ["uniform", "distance"]
    },
    "xgb": {
        "model__n_estimators": [100, 200],
        "model__max_depth": [3, 5],
        "model__learning_rate": [0.05, 0.1]
    }
}

# =====================
# Run GridSearchCV for each model
# =====================
models = {
    "Polynomial Regression": (poly_pipeline, param_grid["poly"]),
    "KNN Regressor": (knn_pipeline, param_grid["knn"]),
    "XGBoost Regressor": (xgb_pipeline, param_grid["xgb"])
}

best_models = {}
for name, (pipe, params) in models.items():
    grid = GridSearchCV(pipe, params, cv=3, scoring="r2", n_jobs=-1)
    grid.fit(X_train, y_train)
    best_models[name] = grid
    print(f"\n{name}")
    print("Best Params:", grid.best_params_)
    print("Best CV RÂ²:", grid.best_score_)

# =====================
# Evaluate Best Models on Test Set
# =====================
for name, grid in best_models.items():
    y_pred = grid.predict(X_test)
    r2 = r2_score(y_test, y_pred)
    print(f"{name} - Test RÂ²: {r2:.4f}")
```
